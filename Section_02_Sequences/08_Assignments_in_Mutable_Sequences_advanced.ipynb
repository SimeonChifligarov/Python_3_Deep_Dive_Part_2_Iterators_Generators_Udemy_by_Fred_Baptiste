## Problem 1 — Basic Slice Replacement Preserves Identity


Predict the output.

The slice has length 3, but the replacement has length 4.


````python
l = [1, 2, 3, 4, 5]
before = id(l)

l[0:3] = ["a", "b", "c", "d"]

after = id(l)

print(l)
print(before == after)
````

### Solution 1


Expected output:

```text
['a', 'b', 'c', 'd', 4, 5]
True
```

Slice assignment mutates the existing list object.

The replacement iterable does not need to have the same length as the original standard slice.


````python
l = [1, 2, 3, 4, 5]
before = id(l)

l[0:3] = ["a", "b", "c", "d"]

after = id(l)

assert l == ["a", "b", "c", "d", 4, 5]
assert before == after

print("All assertions passed.")
````

## Problem 16 — Slice Assignment versus Rebinding


Predict the output.

Compare slice assignment to assigning a new list to the variable.


````python
a = [1, 2, 3]
alias_a = a

a[:] = [10, 20]

b = [1, 2, 3]
alias_b = b

b = [10, 20]

print(a, alias_a, a is alias_a)
print(b, alias_b, b is alias_b)
````

### Solution 16

```text

[10, 20] [1, 2, 3] False
```

````python
a = [1, 2, 3]
alias_a = a
a[:] = [10, 20]

b = [1, 2, 3]
alias_b = b
b = [10, 20]

assert a == [10, 20]
assert alias_a == [10, 20]
assert a is alias_a

assert b == [10, 20]
assert alias_b == [1, 2, 3]
assert b is not alias_b

print("All assertions passed.")
````

## Problem 19 — Immutable Tuple Slice Assignment Error


Predict what happens.


````python
t = (1, 2, 3, 4, 5)

try:
    t[0:3] = (10, 20)
except TypeError as ex:
    print(type(ex).__name__)
    print(str(ex))
````

Expected output includes:


````python
t = (1, 2, 3, 4, 5)

try:
    t[0:3] = (10, 20)
except TypeError as ex:
    error_name = type(ex).__name__
else:
    raise AssertionError("Expected TypeError")

assert error_name == "TypeError"
assert t == (1, 2, 3, 4, 5)

print("All assertions passed.")
````

## Problem 31 — Set as the Right-Hand Side


Predict what is guaranteed and what is not guaranteed.

A set has no reliable insertion order for this purpose.


````python
l = [1, 2, 3, 4, 5]

l[3:3] = {"c", "b", "a"}

print(l)
````

### Solution 31


The exact order of `"a"`, `"b"`, and `"c"` is not something you should rely on.

Guaranteed:

```python
l[:3] == [1, 2, 3]
l[-2:] == [4, 5]
set(l[3:6]) == {"a", "b", "c"}
```

A set is iterable, so slice assignment accepts it, but set iteration order should not be used for logical correctness.


````python
l = [1, 2, 3, 4, 5]

l[3:3] = {"c", "b", "a"}

assert l[:3] == [1, 2, 3]
assert set(l[3:6]) == {"a", "b", "c"}
assert l[-2:] == [4, 5]
assert len(l) == 8

print(l)
print("All assertions passed.")
````

````python
class SimpleSequence:
    def __init__(self, values):
        self._values = list(values)

    def __getitem__(self, index):
        return self._values[index]

    # Implement __setitem__.

    def __repr__(self):
        return f"SimpleSequence({self._values})"


seq = SimpleSequence([1, 2, 3, 4])
seq[1:3] = ["A", "B", "C"]
seq[0] = 99

print(seq)
````

### Solution 49


A wrapper can delegate assignment to the underlying list:

```python
def __setitem__(self, index, value):
    self._values[index] = value
```

The built-in list handles both single indices and slices.


````python
class SimpleSequence:
    def __init__(self, values):
        self._values = list(values)

    def __getitem__(self, index):
        return self._values[index]

    def __setitem__(self, index, value):
        self._values[index] = value

    def __repr__(self):
        return f"SimpleSequence({self._values})"


seq = SimpleSequence([1, 2, 3, 4])
seq[1:3] = ["A", "B", "C"]
seq[0] = 99

assert seq._values == [99, "A", "B", "C", 4]

print(seq)
print("All assertions passed.")
````

# Key Takeaways

1. Mutable sequences such as lists support item and slice assignment.
2. Standard slice assignment can replace, insert, or delete elements.
3. Standard slice replacement does not require equal lengths.
4. Assigning to an empty standard slice inserts elements.
5. Assigning an empty iterable to a standard slice deletes elements.
6. Slice assignment mutates the existing list object and preserves identity.
7. Aliases observe mutations caused by slice assignment.
8. `l[:] = new_values` replaces all contents in place.
9. `l = new_values` rebinds the variable and does not update aliases.
10. The right-hand side of slice assignment must be iterable.
11. Strings are iterable, so assigning a string to a slice inserts characters.
12. Sets are iterable, but their iteration order should not be relied on.
13. Immutable sequences such as tuples and strings do not support slice assignment.
14. Extended slice assignment requires the replacement iterable to have exactly the same length as the selected slice.
15. `l[::2] = []` fails when the extended slice is non-empty, but `del l[::2]` can delete those elements.
16. Negative-step slices assign values in the selected index order.
17. Slice assignment is a powerful tool, but it can easily introduce aliasing bugs.
18. Use copying when you do not want to mutate caller-owned data.
